```python
import sqlite3 as sql
con=sql.connect(database="inventorydb.sqlite")
cursor=con.cursor()
try:
    cursor.execute("create table students(stu_id int,stu_name text,stu_mob text,stu_course text,course_fee int,bal int)")
    cursor.execute("create table course(course_name text primary key,course_fee int)")

except Exception as e:
    print(e)
con.commit()
con.close()
```

```python
from tkinter import *
from tkinter import messagebox
from tkinter.ttk import Combobox
win=Tk()
win.state('zoomed')
win.resizable(width=False,height=False)
header_frm=Frame(win)
header_frm.configure(bg='white')
header_frm.place(x=0,y=0,relwidth=1,relheight=.2)

title_lbl=Label(header_frm,text="Inventroy Management",bg='white',font=('',50,'bold','underline'))
title_lbl.pack()

def main_body():
    def reset():
        user_entry.delete(0,"end")
        pass_entry.delete(0,"end")
        user_entry.focus()
    
    def login():
        u=user_entry.get()
        p=pass_entry.get()
        if(len(u)==0 or len(p)==0):
            messagebox.showwarning('validation',"Username/Password can't be empty")
        else:
            if(u=='admin' and p=='admin'):
                frm.destroy()
                login_body()
            else:
                messagebox.showerror('validation',"Invalid Username/Password")
    frm=Frame(win)
    frm.configure(bg='grey')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    user_lbl=Label(frm,text="Username:",font=('',25,'bold'),bg='red')
    user_lbl.place(relx=.3,rely=.2)

    user_entry=Entry(frm,font=('','20'),bd=7)
    user_entry.focus()
    user_entry.place(relx=.45,rely=.2)
    
    pass_lbl=Label(frm,text="Password:",font=('',25,'bold'),bg='green')
    pass_lbl.place(relx=.3,rely=.35)

    pass_entry=Entry(frm,font=('','20'),bd=7,show='*')
    pass_entry.place(relx=.45,rely=.35)
    
    login_btn=Button(frm,text="login",font=('','20'),bd=7,command=login)
    login_btn.place(relx=.4,rely=.5)
    
    reset_btn=Button(frm,text="reset",font=('','20'),bd=7,command=reset)
    reset_btn.place(relx=.5,rely=.5)

    
def login_body():
    def logout():
        frm.destroy()
        main_body()
        
    def register_student():
        frm.destroy()
        register_student_body()
        
    def find():
        frm.destroy()
        find_student_body()
    
    def deposit():
        frm.destroy()
        deposit_fee_body()
    
    def add():
        frm.destroy()
        add_course_body()
        
        
    frm=Frame(win)
    frm.configure(bg='grey')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    wel_lbl=Label(frm,text="Welcome,admin",font=('',20),bg='grey')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('',20),bd=7,command=logout)
    logout_btn.place(relx=.92,rely=0)
    
    reg_btn=Button(frm,text="Register Inventory",font=('',20),bd=7,width=15,command=register_student)
    reg_btn.place(relx=.4,rely=.1)
    
    find_btn=Button(frm,text="Find ",font=('',20),bd=7,width=15,command=find)
    find_btn.place(relx=.4,rely=.3)
    
    dep_btn=Button(frm,text="Deposit amount",font=('',20),bd=7,width=15,command=deposit)
    dep_btn.place(relx=.4,rely=.5)
    
    course_btn=Button(frm,text="Add Inventory",font=('',20),bd=7,width=15,command=add)
    course_btn.place(relx=.4,rely=.7)
    
def register_student_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
        
    def register_stu_db():
        stu_id=id_entry.get()
        stu_name=name_entry.get()
        stu_mob=mob_entry.get()
        stu_crs=course_list.get()
        stu_fee=fee_entry.get()
        crs,crs_fee=stu_crs.split()
        con=sql.connect(database="inventorydb.sqlite")
        cursor=con.cursor()
        cursor.execute("insert into students values(?,?,?,?,?,?)",(stu_id,stu_name,stu_mob,crs,crs_fee,(int(crs_fee)-int(stu_fee))))
        con.commit()
        con.close()
        messagebox.showinfo('Registration',"Inventory Registered")
        frm.destroy()
        register_student_body()
       
    frm=Frame(win)
    frm.configure(bg='grey')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    wel_lbl=Label(frm,text="Welcome,admin",font=('',20),bg='grey')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('',20),bd=7,command=logout)
    logout_btn.place(relx=.92,rely=0)
    
    home_btn=Button(frm,text="home",font=('',20),bd=7,command=home)
    home_btn.place(relx=.82,rely=0)
    
    con=sql.connect(database="inventorydb.sqlite")
    cursor=con.cursor()
    cursor.execute("select max(stu_id) from students")
    max_id=cursor.fetchone()
    if(max_id[0]==None):
        stu_id_db=1
    else:
        stu_id_db=max_id[0]+1
    con.close()
    stu_id=Label(frm,text="Customer Id:",font=('',20,'bold'),bg='red')
    stu_id.place(relx=.3,rely=.01)
    
    id_entry=Entry(frm,font=('','20'),bd=7)
    id_entry.place(relx=.47,rely=.01)
    id_entry.insert(0,str(stu_id_db))
    id_entry.configure(state='disabled')
    
    stu_name=Label(frm,text="Customer Name:",font=('',20,'bold'),bg='red')
    stu_name.place(relx=.3,rely=.15)
    
    name_entry=Entry(frm,font=('','20'),bd=7)
    name_entry.focus()
    name_entry.place(relx=.47,rely=.15)
    
    stu_mob=Label(frm,text="Customer Mob:",font=('',20,'bold'),bg='red')
    stu_mob.place(relx=.3,rely=.3)
    
    mob_entry=Entry(frm,font=('','20'),bd=7)
    mob_entry.place(relx=.47,rely=.3)
    
    stu_course=Label(frm,text="Select Inventory:",font=('',20,'bold'),bg='red')
    stu_course.place(relx=.3,rely=.45)
    
    con=sql.connect(database="inventorydb.sqlite")
    cursor=con.cursor()
    cursor.execute("select * from course")
    courses=cursor.fetchall()
    courses.insert(0,"---select----")
    con.close()
    course_list=Combobox(frm,values=courses,font=('',20,'bold'))
    course_list.place(relx=.47,rely=.45)
    course_list.current(0)
    
    reg_fee=Label(frm,text="Reg. Amt:",font=('',20,'bold'),bg='red')
    reg_fee.place(relx=.3,rely=.6)
    
    fee_entry=Entry(frm,font=('','20'),bd=7)
    fee_entry.place(relx=.47,rely=.6)
    
    reg_btn=Button(frm,text="Register",font=('',20),bd=7,command=register_stu_db)
    reg_btn.place(relx=.52,rely=.75)
    
def find_student_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
    
    def search():
        stu_id=id_entry.get()
        con=sql.connect(database="inventorydb.sqlite")
        cursor=con.cursor()
        cursor.execute("select * from students where stu_id=?",(stu_id,))
        row=cursor.fetchone()
        if(row==None):
            messagebox.showwarning("Search","Customer does not exist")
        else:
            messagebox.showinfo("Search",str(row))
    frm=Frame(win)
    frm.configure(bg='grey')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    wel_lbl=Label(frm,text="Welcome,admin",font=('',20),bg='red')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('',20),bd=7,command=logout)
    logout_btn.place(relx=.92,rely=0)
    
    home_btn=Button(frm,text="home",font=('',20),bd=7,command=home)
    home_btn.place(relx=.82,rely=0)
    
    stu_id=Label(frm,text="Customer ID:",font=('',20,'bold'),bg='red')
    stu_id.place(relx=.3,rely=.15)
    
    id_entry=Entry(frm,font=('','20'),bd=7)
    id_entry.focus()
    id_entry.place(relx=.47,rely=.15)
    
    search_btn=Button(frm,text="Search",font=('',20),bd=7,command=search)
    search_btn.place(relx=.4,rely=.3)


def deposit_fee_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
        
    def fetch_fee(event):
        stu_id=id_entry.get()
        con=sql.connect(database="inventorydb.sqlite")
        cursor=con.cursor()
        cursor.execute("select bal from students where stu_id=?",(stu_id,))
        row=cursor.fetchone()
        if(row==None):
            messagebox.showwarning("","customer does not exist")
        else:
            fee_entry.configure(state="normal")
            fee_entry.delete(0,"end")
            fee_entry.insert(0,row[0])
            fee_entry.configure(state="disabled")
            dep_entry.focus()
    def deposit():
        stu_id=id_entry.get()
        dep_fee=dep_entry.get()
        con=sql.connect(database="inventorydb.sqlite")
        cursor=con.cursor()
        cursor.execute("update students set bal=bal-? where stu_id=?",(int(dep_fee),stu_id))
        con.commit()
        con.close()
        messagebox.showinfo('','amount deposited')
    frm=Frame(win)
    frm.configure(bg='red')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    wel_lbl=Label(frm,text="Welcome,admin",font=('',20),bg='red')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('',20),bd=7,command=logout)
    logout_btn.place(relx=.92,rely=0)
    
    home_btn=Button(frm,text="home",font=('',20),bd=7,command=home)
    home_btn.place(relx=.82,rely=0)
    
    stu_id=Label(frm,text="Customer id:",font=('',20,'bold'),bg='red')
    stu_id.place(relx=.3,rely=.15)
    
    id_entry=Entry(frm,font=('','20'),bd=7)
    id_entry.focus()
    id_entry.place(relx=.47,rely=.15)
    id_entry.bind("<FocusOut>",fetch_fee)
    
    stu_fee=Label(frm,text="Remaining Amount:",font=('',20,'bold'),bg='red')
    stu_fee.place(relx=.3,rely=.3)
    
    fee_entry=Entry(frm,font=('','20'),bd=7,state='disabled')
    fee_entry.place(relx=.47,rely=.3)
    
    dep_fee=Label(frm,text="Amount:",font=('',20,'bold'),bg='red')
    dep_fee.place(relx=.3,rely=.45)
    
    dep_entry=Entry(frm,font=('','20'),bd=7)
    dep_entry.place(relx=.47,rely=.45)
    
    dep_btn=Button(frm,text="Deposit",font=('',20),bd=7,command=deposit)
    dep_btn.place(relx=.4,rely=.7)
    
def add_course_body():
    def logout():
        frm.destroy()
        main_body()
    def home():
        frm.destroy()
        login_body()
        
    def add_course_db():
        n=name_entry.get()
        n=n.upper()
        f=fee_entry.get()
        con=sql.connect(database="inventorydb.sqlite")
        cursor=con.cursor()
        try:
            cursor.execute("insert into course values(?,?)",(n,f))
            con.commit()
            messagebox.showinfo('','Inventory Added')
        except Exception as e:
            messagebox.showerror('',str(e))
        
        con.close()
        
    frm=Frame(win)
    frm.configure(bg='grey')
    frm.place(x=0,rely=.2,relwidth=1,relheight=.8)
    wel_lbl=Label(frm,text="Welcome,admin",font=('',20),bg='red')
    wel_lbl.place(relx=0,rely=0)
    
    logout_btn=Button(frm,text="logout",font=('',20),bd=7,command=logout)
    logout_btn.place(relx=.92,rely=0)
    
    home_btn=Button(frm,text="home",font=('',20),bd=7,command=home)
    home_btn.place(relx=.82,rely=0)
    
    course_name=Label(frm,text="Inventory Name:",font=('',20,'bold'),bg='red')
    course_name.place(relx=.3,rely=.15)
    
    name_entry=Entry(frm,font=('','20'),bd=7)
    name_entry.focus()
    name_entry.place(relx=.47,rely=.15)
    
    course_fee=Label(frm,text="Inventory Amount:",font=('',20,'bold'),bg='red')
    course_fee.place(relx=.3,rely=.3)
    
    fee_entry=Entry(frm,font=('','20'),bd=7)
    fee_entry.focus()
    fee_entry.place(relx=.47,rely=.3)
    
    add_btn=Button(frm,text="Add",font=('',20),bd=7,command=add_course_db)
    add_btn.place(relx=.4,rely=.45)

    
main_body()
win.mainloop()
```
